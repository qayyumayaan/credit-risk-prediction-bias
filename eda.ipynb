Data Analysis:

Distribution of Target Variable:

```python
import pandas as pd
import matplotlib.pyplot as plt

# TARGET legend mapping
target_map = {0: "Good Borrower", 1: "Bad Borrower"}

# Count of each class
target_counts = train['TARGET'].value_counts().sort_index()  # 0 then 1
labels = [target_map[idx] for idx in target_counts.index]

plt.figure(figsize=(6,4))

# Bar plot
bars = plt.bar(labels, target_counts.values)

plt.title("Borrower Type Distribution")
plt.xlabel("Borrower Type")
plt.ylabel("Number of Clients")

# Optional: annotate values on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height,
             f'{height:,}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

# If you also want a printed "legend table" in text:
legend_table = pd.DataFrame({
    "TARGET": [0, 1],
    "Label": ["Good Borrower", "Bad Borrower"]
})
print(legend_table)
```

Q1) Does car ownership reduce or increase the chance of loan default?

```python
# Map Y/N to Yes/No
train['CAR_YES_NO'] = train['FLAG_OWN_CAR'].map({'Y': 'Yes', 'N': 'No'})

# Default rate by car ownership (percentage)
default_rates = train.groupby('CAR_YES_NO')['TARGET'].mean() * 100

plt.figure(figsize=(6,4))
default_rates.plot(kind='bar')
plt.title("Default Rate by Car Ownership")
plt.xlabel("Owns Car")
plt.ylabel("Default Rate (%)")
plt.tight_layout()
plt.show()
```

Q2) Are car owners richer than non–car owners?

```python
# Map Y/N to Yes/No
train['CAR_YES_NO'] = train['FLAG_OWN_CAR'].map({'Y': 'Yes', 'N': 'No'})

# Income distribution boxplot
plt.figure(figsize=(6,4))
train.boxplot(column='AMT_INCOME_TOTAL', by='CAR_YES_NO')
plt.title("Income Distribution by Car Ownership")
plt.suptitle("")  # remove the auto-title
plt.xlabel("Owns Car")
plt.ylabel("Income (Annual)")
plt.tight_layout()
plt.show()
```

Q3) Education vs Default Rate:

```python
# Default rate by education level (percentage)
edu_default_pct = train.groupby('NAME_EDUCATION_TYPE')['TARGET'].mean() * 100

edu_default_pct.sort_values().plot(kind='barh')

plt.title("Default Rate by Education Level")
plt.xlabel("Default Rate (%)")
plt.show()
```

Q4) Who defaults more by income type?

```python
# Default rate by income type (percentage)
risk = train.groupby('NAME_INCOME_TYPE')['TARGET'].mean() * 100

# Drop Student and Businessman categories
risk = risk.drop(['Student', 'Businessman', 'Maternity leave'], errors='ignore')

plt.figure(figsize=(7,7))
plt.pie(
    risk,
    labels=risk.index,
    autopct='%1.1f%%',
    startangle=140
)

plt.title("Default Rate (%) by Income Type")
plt.axis('equal')
plt.show()
```

Q5) Age vs Default

```python
# Create AGE column
train['AGE'] = (-train['DAYS_BIRTH']) / 365

# Determine min and max age
min_age = int(train['AGE'].min())
max_age = int(train['AGE'].max())

# Create bins of width = 8 years
bins = np.arange(min_age, max_age + 8, 8)

# Categorize into 8-year intervals
age_bins = pd.cut(train['AGE'], bins=bins)

# Default rate (%) for each age interval
age_default = train.groupby(age_bins)['TARGET'].mean() * 100

# Clean interval labels (e.g., "20–28")
clean_labels = [
    f"{int(interval.left)}–{int(interval.right)}"
    for interval in age_default.index
]

# Plot
plt.figure(figsize=(10, 6))
plt.bar(clean_labels, age_default.values)

plt.title("Default Rate (%) by Age Range (8-year intervals)")
plt.xlabel("Age Range (years)")
plt.ylabel("Default Rate (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
