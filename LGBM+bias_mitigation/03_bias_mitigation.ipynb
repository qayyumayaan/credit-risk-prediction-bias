```python
# Function to calculate missing values by column# Funct 
def missing_values_table(df):
        # Total missing values
        mis_val = df.isnull().sum()
        
        # Percentage of missing values
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        
        # Make a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        # Rename the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        
        # Sort the table by percentage of missing descending
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        
        # Print some summary information
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        
        # Return the dataframe with missing information
        return mis_val_table_ren_columns

# Missing values statistics
missing_values = missing_values_table(train)
missing_values.head(20)
```

```python
def preprocess_and_encode(train_df, test_df, target_col="TARGET"):
    """
    Returns:
        X_enc, X_test_enc, y, feature_names, ct
    """
    import pandas as pd
    from sklearn.compose import ColumnTransformer
    from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

    # 1. Split target
    y = train_df[target_col]
    train_X = train_df.drop(columns=[target_col])

    # 2. Align train/test
    train_X, test_X = train_X.align(test_df, join="inner", axis=1)

    # 3. Identify categorical columns
    one_hot_cols = [
        'NAME_CONTRACT_TYPE', 'CODE_GENDER', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY',
        'NAME_TYPE_SUITE', 'NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE',
        'NAME_FAMILY_STATUS', 'NAME_HOUSING_TYPE', 'OCCUPATION_TYPE',
        'ORGANIZATION_TYPE', 'FONDKAPREMONT_MODE', 'HOUSETYPE_MODE',
        'EMERGENCYSTATE_MODE'
    ]

    # Keep only those present
    one_hot_cols = [c for c in one_hot_cols if c in train_X.columns]

    # 4. Split binary vs multi
    binary_cats = []
    multi_cats = []

    for col in one_hot_cols:
        unique_vals = train_X[col].dropna().unique()
        if len(unique_vals) <= 2:
            binary_cats.append(col)
        else:
            multi_cats.append(col)

    # 5. Build ColumnTransformer
    ct = ColumnTransformer(
        transformers=[
            ("binary", OrdinalEncoder(), binary_cats),
            ("onehot", OneHotEncoder(handle_unknown='ignore', sparse_output=False), multi_cats),
        ],
        remainder="passthrough"
    )

    # 6. Fit/transform
    X_enc = ct.fit_transform(train_X)
    X_test_enc = ct.transform(test_X)

    feature_names = ct.get_feature_names_out()

    return X_enc, X_test_enc, y, feature_names, ct

def make_numeric_for_corr(df):
    df_corr = df.copy()
    for col in df_corr.columns:
        # Convert objects/categories to integer codes for correlation
        if df_corr[col].dtype == 'object' or str(df_corr[col].dtype).startswith('category'):
            df_corr[col] = df_corr[col].astype('category').cat.codes
    return df_corr

```

So there are a lot of features that are correlated with gender and age. However, if we were to drop all of these features, our model would not be very accurate. To solve this problem, we will run an overlap test to identify features that can cause a positivity violation with our protected classes of gender and age. 

```python
#  0. Convert categorical to numeric for correlation & overlap 
train_corr = make_numeric_for_corr(train)

#  1. Define protected variables 
biased_cols = ['CODE_GENDER', 'DAYS_BIRTH']   # core sensitive features

#  2. Create Gender Binary (0 = M, 1 = F) 
train['_GENDER_BIN'] = (train['CODE_GENDER'] == 'F').astype(int)

#  3. Create Age Group Bins for positivity testing 
# DAYS_BIRTH is negative in Home Credit, so invert it:
train['_AGE'] = -train['DAYS_BIRTH']

# Use quartiles for overlap testing
train['_AGE_BIN'] = pd.qcut(train['_AGE'], q=4, labels=False, duplicates='drop')

#  4. Define candidate (non-protected) features 
protected_vars = biased_cols + ['_GENDER_BIN', '_AGE_BIN', '_AGE']

candidate_features = [
    c for c in train.columns 
    if c not in protected_vars and c not in ['TARGET']
]

print("Number of candidate features:", len(candidate_features))


#  5. POSITIVITY TEST FOR GENDER (M vs F) 

female_unique = train.loc[train['_GENDER_BIN'] == 1, candidate_features].nunique()
viol_female = list(female_unique[female_unique < 2].index)

male_unique   = train.loc[train['_GENDER_BIN'] == 0, candidate_features].nunique()
viol_male = list(male_unique[male_unique < 2].index)

print("\n Gender Positivity Violations ")
print("Violates for FEMALES:", viol_female)
print("Violates for MALES:", viol_male)


#  6. POSITIVITY TEST FOR AGE (each age bin) 

viol_age_bins = {}

for age_group in sorted(train['_AGE_BIN'].dropna().unique()):
    gmask = train['_AGE_BIN'] == age_group
    nunique_vals = train.loc[gmask, candidate_features].nunique()
    viol = list(nunique_vals[nunique_vals < 2].index)
    viol_age_bins[int(age_group)] = viol

print("\n Age Positivity Violations ")
for k, v in viol_age_bins.items():
    print(f"Age bin {k}: {v}")


#  7. Combine all violations 

positivity_violations = set(viol_female) | set(viol_male)

for vlist in viol_age_bins.values():
    positivity_violations |= set(vlist)

positivity_violations = sorted(positivity_violations)

print("\n ALL Positivity Violations Across Gender + Age ")
print(positivity_violations)
```

This is a very interesting result. The test we just ran studied the dataset for positivity (overlap) violations. In the certain protected groups, these two features showed no variability. For females for example, the FLAG_MOBIL variable always has the same result. This flag also presents itself as an age-related bias. With a lack of variability, these features still reinforce structural bias and undermine the fairness adjusted modeling technique. 

```python
# Last step: Drop all the biased columns from the unbiased set

all_biased_to_drop = set(biased_cols) | set(proxy_features) | set(positivity_violations)

print("Dropping these biased features:", all_biased_to_drop)

train_unbiased = train.drop(columns=all_biased_to_drop)
test_unbiased = test.drop(columns=all_biased_to_drop)

print(f"Dropped {len(all_biased_to_drop)} features. This is {(len(all_biased_to_drop) / len(train_biased.columns)) * 100:.2f}% of the original.")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix, roc_auc_score
)
import seaborn as sns
from sklearn.model_selection import cross_val_predict

def scores(y_true, y_pred, y_pred_proba, title="Model Evaluation"):
    # Compute metrics
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    auc = roc_auc_score(y_true, y_pred_proba)
    cm = confusion_matrix(y_true, y_pred)

    # Print metric summary
    print(f"{title}")
    print("Accuracy:", acc)
    print("Precision:", prec)
    print("Recall:", rec)
    print("F1:", f1)
    print("ROC-AUC:", auc)
    print("\nConfusion Matrix (Raw Values):\n", cm)
    
    # Plot confusion matrix
    plt.figure(figsize=(6, 5))
    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues",
        xticklabels=["Pred 0", "Pred 1"],
        yticklabels=["True 0", "True 1"]
    )
    plt.title(f"Confusion Matrix: {title}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()
```

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import cross_val_predict
import pandas as pd

# Shared LightGBM configuration
lgbm_params = dict(
    objective="binary",
    metric="auc",
    boosting_type="gbdt",
    n_estimators=800,
    learning_rate=0.02,
    num_leaves=31,
    subsample=0.8,
    colsample_bytree=0.8,
    class_weight="balanced",
    random_state=42,
    verbose=-1,
)

# Dataset registry
datasets = {
    "unbiased": {
        "X_train": X_unbiased_enc,
        "y_train": y_unbiased,
        "X_test": X_test_unbiased_enc,
        "feature_names": feature_names_unbiased,
    },
    "biased": {
        "X_train": X_biased_enc,
        "y_train": y_biased,
        "X_test": X_test_biased_enc,
        "feature_names": feature_names_biased,
    },
}

# Containers for results
cv_results = {}
test_predictions = {}
feature_importance = {}
models = {}

# Train + evaluate both datasets
for name, data in datasets.items():
    print(f"\nLightGBM ({name.upper()})")

    model = LGBMClassifier(**lgbm_params)

    # Cross-validated predictions
    y_pred = cross_val_predict(
        model,
        data["X_train"],
        data["y_train"],
        cv=5,
        method="predict",
    )

    y_pred_proba = cross_val_predict(
        model,
        data["X_train"],
        data["y_train"],
        cv=5,
        method="predict_proba",
    )[:, 1]

    print("\nLightGBM CV Scores")
    scores(data["y_train"], y_pred, y_pred_proba, title=f"Model Evaluation ({name})")

    # Store CV results
    cv_results[name] = {
        "y_pred": y_pred,
        "y_pred_proba": y_pred_proba,
    }

    # Fit final model on full training data
    model.fit(data["X_train"], data["y_train"])
    models[name] = model

    # Test predictions
    test_predictions[name] = model.predict_proba(data["X_test"])[:, 1]

    # Feature importance
    fi = pd.Series(
        model.feature_importances_,
        index=data["feature_names"],
    ).sort_values(ascending=False)

    feature_importance[name] = fi

    print("\nTop 20 LightGBM Features:")
    print(fi.head(20))
```

I want to answer this question: How would the biased model behave if it were forced to act like the unbiased model, and which biased features prevent that?

```python
import shap
import numpy as np
import pandas as pd

# Use TreeExplainer (exact for LightGBM)
explainer_biased = shap.TreeExplainer(models["biased"])
explainer_unbiased = shap.TreeExplainer(models["unbiased"])

# Compute SHAP values on TRAINING data
# (use training for attribution; test for validation)
shap_biased = explainer_biased.shap_values(X_biased_enc)
shap_unbiased = explainer_unbiased.shap_values(X_unbiased_enc)

# Binary classification: take class-1 SHAP values if returned as list
if isinstance(shap_biased, list):
    shap_biased = shap_biased[1]
if isinstance(shap_unbiased, list):
    shap_unbiased = shap_unbiased[1]


# Feature name sets
biased_features = feature_names_biased
unbiased_features = feature_names_unbiased

biased_set = set(biased_features)
unbiased_set = set(unbiased_features)

shared_features = sorted(biased_set & unbiased_set)
biased_only_features = sorted(biased_set - unbiased_set)

print(f"Shared features: {len(shared_features)}")
print(f"Biased-only features: {len(biased_only_features)}")


# Index maps
biased_idx = {f: i for i, f in enumerate(biased_features)}
unbiased_idx = {f: i for i, f in enumerate(unbiased_features)}

# SHAP difference for shared features
delta_shap_shared = np.column_stack([
    shap_biased[:, biased_idx[f]] - shap_unbiased[:, unbiased_idx[f]]
    for f in shared_features
])

delta_shap_shared_df = pd.DataFrame(
    delta_shap_shared,
    columns=shared_features
)


# Pure bias vector
delta_shap_biased_only = np.column_stack([
    shap_biased[:, biased_idx[f]]
    for f in biased_only_features
])

delta_shap_biased_only_df = pd.DataFrame(
    delta_shap_biased_only,
    columns=biased_only_features
)


# Mean absolute contribution (robust global importance)
bias_attribution = pd.concat([
    delta_shap_shared_df.abs().mean().rename("shared_delta"),
    delta_shap_biased_only_df.abs().mean().rename("biased_only")
], axis=1).fillna(0)

bias_attribution["total_bias_effect"] = (
    bias_attribution["shared_delta"] + bias_attribution["biased_only"]
)

bias_attribution = bias_attribution.sort_values(
    "total_bias_effect", ascending=False
)

print("Top bias-driving features:")
print(bias_attribution.head(20))


# Validate against prediction deltas
# Prediction delta
delta_pred = (
    cv_results["biased"]["y_pred_proba"]
    - cv_results["unbiased"]["y_pred_proba"]
)

# Total SHAP delta per sample
total_shap_delta = (
    delta_shap_shared_df.sum(axis=1)
    + delta_shap_biased_only_df.sum(axis=1)
)

corr = np.corrcoef(delta_pred, total_shap_delta)[0, 1]
print(f"Correlation(prediction , SHAP ): {corr:.3f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Helper: safe quantiles for large arrays

def _winsorize(x, lo=0.01, hi=0.99):
    qlo, qhi = np.quantile(x, [lo, hi])
    return np.clip(x, qlo, qhi)

# 1) VISUALIZE RESULTS

# 1A) Sanity check: prediction delta vs SHAP delta scatter
plt.figure(figsize=(6, 5))
x = _winsorize(delta_pred, 0.01, 0.99)
y = _winsorize(total_shap_delta.values if hasattr(total_shap_delta, "values") else total_shap_delta, 0.01, 0.99)

plt.scatter(x, y, s=3, alpha=0.15)
plt.axhline(0, color="gray", linewidth=1, alpha=0.5)
plt.axvline(0, color="gray", linewidth=1, alpha=0.5)
plt.xlabel(" prediction = p_biased - p_unbiased (winsorized)")
plt.ylabel(" SHAP total (winsorized)")
plt.title(f"Consistency: Prediction  vs SHAP  (corr={corr:.3f})")
plt.tight_layout()
plt.show()

# 1B) Distribution of prediction deltas
plt.figure(figsize=(6, 4))
sns.histplot(delta_pred, bins=60, kde=True)
plt.axvline(0, color="gray", linewidth=1)
plt.title("Distribution of Prediction Shift: p_biased - p_unbiased")
plt.xlabel(" prediction")
plt.tight_layout()
plt.show()

# 1C) Top-K bias features barplot (global)
TOPK = 25
topk = bias_attribution.head(TOPK).copy()

plt.figure(figsize=(9, 6))
topk_plot = topk.sort_values("total_bias_effect", ascending=True)
plt.barh(topk_plot.index, topk_plot["total_bias_effect"], color="steelblue", alpha=0.9)
plt.title(f"Top {TOPK} Features Driving Bias (|SHAP| aggregated)")
plt.xlabel("Total bias effect = shared_delta + biased_only")
plt.tight_layout()
plt.show()

# 1D) Decompose: shared vs biased-only for the same top-K
plt.figure(figsize=(9, 6))
topk_plot = topk.sort_values("total_bias_effect", ascending=True)
plt.barh(topk_plot.index, topk_plot["shared_delta"], label="shared_delta", alpha=0.8)
plt.barh(topk_plot.index, topk_plot["biased_only"], left=topk_plot["shared_delta"], label="biased_only", alpha=0.8)
plt.title(f"Top {TOPK} Bias Features: Shared  vs Biased-only")
plt.xlabel("Contribution (mean |SHAP|)")
plt.legend()
plt.tight_layout()
plt.show()

# 2) BIAS ATTRIBUTION BY OUTCOME (TP / FP / FN / TN)

# Use out-of-fold biased model predictions to define outcomes
y_true = y_biased.values if hasattr(y_biased, "values") else np.asarray(y_biased)
p_b = cv_results["biased"]["y_pred_proba"]
yhat_b = (p_b >= 0.5).astype(int)

# Confusion strata
mask_tp = (y_true == 1) & (yhat_b == 1)
mask_fp = (y_true == 0) & (yhat_b == 1)
mask_fn = (y_true == 1) & (yhat_b == 0)
mask_tn = (y_true == 0) & (yhat_b == 0)

outcome_masks = {
    "TP": mask_tp,
    "FP": mask_fp,
    "FN": mask_fn,
    "TN": mask_tn,
}

print("Counts by outcome (biased model @ 0.5 threshold):")
for k, m in outcome_masks.items():
    print(f"  {k}: {int(m.sum())}")

# Function to compute top features for a subset
def top_bias_features_for_mask(mask, topk=20):
    shared = delta_shap_shared_df.loc[mask].abs().mean()
    biased_only = delta_shap_biased_only_df.loc[mask].abs().mean()
    df = pd.concat([shared.rename("shared_delta"), biased_only.rename("biased_only")], axis=1).fillna(0)
    df["total_bias_effect"] = df["shared_delta"] + df["biased_only"]
    return df.sort_values("total_bias_effect", ascending=False).head(topk)

# 2A) Print top features by outcome
TOPK_OUTCOME = 15
top_by_outcome = {}
for outcome, mask in outcome_masks.items():
    top_by_outcome[outcome] = top_bias_features_for_mask(mask, topk=TOPK_OUTCOME)
    print(f"\nTop {TOPK_OUTCOME} bias-driving features for outcome={outcome}:")
    display(top_by_outcome[outcome])

# 2B) Compare outcomes on a common set of features (union of top features)
union_feats = sorted(set().union(*[df.index.tolist() for df in top_by_outcome.values()]))

heat = pd.DataFrame(index=union_feats)
for outcome, df in top_by_outcome.items():
    heat[outcome] = df.reindex(union_feats)["total_bias_effect"].fillna(0)

plt.figure(figsize=(10, max(6, 0.25 * len(union_feats))))
sns.heatmap(heat, cmap="viridis", linewidths=0.3)
plt.title("Bias Attribution by Outcome (mean |SHAP| per feature)")
plt.xlabel("Outcome (biased model)")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()

# 2C) Outcome-level summary of how bias shifts predictions
outcome_summary = []
for outcome, mask in outcome_masks.items():
    outcome_summary.append({
        "outcome": outcome,
        "n": int(mask.sum()),
        "mean_delta_pred": float(np.mean(delta_pred[mask])) if mask.sum() else np.nan,
        "median_delta_pred": float(np.median(delta_pred[mask])) if mask.sum() else np.nan,
        "mean_abs_delta_pred": float(np.mean(np.abs(delta_pred[mask]))) if mask.sum() else np.nan,
    })
outcome_summary_df = pd.DataFrame(outcome_summary).set_index("outcome")
display(outcome_summary_df)

plt.figure(figsize=(6, 4))
sns.barplot(x=outcome_summary_df.index, y=outcome_summary_df["mean_delta_pred"])
plt.axhline(0, color="gray", linewidth=1)
plt.title("Mean Prediction Shift (p_biased - p_unbiased) by Outcome")
plt.ylabel("Mean  prediction")
plt.tight_layout()
plt.show()

# 3) CAUSAL AUDIT REPORT (structured output)

# The report below is a governance-friendly summary.
# It does NOT claim causal identification; it frames results as "incremental contribution conditional on control model".

def causal_audit_report(topk_global=15, topk_outcome=10):
    # Global top-k
    global_top = bias_attribution.head(topk_global).copy()

    # Basic shift stats
    dp = np.asarray(delta_pred)
    report = {}
    report["n_samples"] = int(len(dp))
    report["delta_pred_mean"] = float(np.mean(dp))
    report["delta_pred_median"] = float(np.median(dp))
    report["delta_pred_std"] = float(np.std(dp))
    report["delta_pred_p95"] = float(np.quantile(dp, 0.95))
    report["delta_pred_p05"] = float(np.quantile(dp, 0.05))
    report["shap_delta_corr"] = float(corr)

    # Outcome summaries
    outcome_rows = []
    for outcome, mask in outcome_masks.items():
        if mask.sum() == 0:
            continue
        outcome_rows.append({
            "outcome": outcome,
            "n": int(mask.sum()),
            "mean_delta_pred": float(np.mean(dp[mask])),
            "mean_abs_delta_pred": float(np.mean(np.abs(dp[mask]))),
        })
    outcome_df = pd.DataFrame(outcome_rows).set_index("outcome").sort_index()

    # Outcome-specific top-k tables
    outcome_tables = {}
    for outcome, mask in outcome_masks.items():
        if mask.sum() == 0:
            continue
        outcome_tables[outcome] = top_bias_features_for_mask(mask, topk=topk_outcome)

    return report, global_top, outcome_df, outcome_tables

report_stats, global_top_tbl, outcome_stats_tbl, outcome_top_tbls = causal_audit_report(
    topk_global=20,
    topk_outcome=12
)

print("\n" + "="*70)
print("CAUSAL AUDIT REPORT (Incremental Attribution; Conditional on Control Model)")
print("="*70)

print("\nScope & Setup")
print("- Models compared: LightGBM(biased features) vs LightGBM(unbiased features)")
print("- Attribution method: SHAP = SHAP_biased - SHAP_unbiased (shared features),")
print("  plus SHAP_biased for biased-only features.")
print("- Interpretation: Incremental contribution conditional on the control model (unbiased).")
print("- Note: This is not a formal causal identification claim; it is a structured counterfactual attribution.")

print("\nShift Summary (p_biased - p_unbiased)")
for k, v in report_stats.items():
    print(f"- {k}: {v}")

print("\nGlobal Bias Drivers (Top features by mean |SHAP|)")
display(global_top_tbl)

print("\nOutcome-Level Shift Summary (biased model outcomes @ threshold 0.5)")
display(outcome_stats_tbl)

print("\nOutcome-Level Bias Drivers (Top features by mean |SHAP|)")
for outcome, tbl in outcome_top_tbls.items():
    print(f"\nOutcome = {outcome}")
    display(tbl)

print("\nAudit Recommendations (Actionable)")
print("1) Inspect the top biased-only features and confirm whether they are proxies for protected attributes.")
print("2) For features dominant in FP vs FN strata, evaluate policy harm: FP-heavy bias suggests over-rejection.")
print("3) Stress-test by removing top-k biased-only features and re-evaluating AUC + calibration + error parity.")
print("4) If you have a protected attribute label (or proxy), compute  prediction shift parity across groups.")
print("5) Consider constrained training (monotonic constraints / adversarial debiasing / feature sanitization) on the top drivers.")

# Optional: export tables for governance artifacts
global_top_tbl.to_csv("bias_audit_global_top_features.csv")
outcome_stats_tbl.to_csv("bias_audit_outcome_shift_summary.csv")
for outcome, tbl in outcome_top_tbls.items():
    tbl.to_csv(f"bias_audit_top_features_{outcome}.csv")

print("\nSaved:")
print("- bias_audit_global_top_features.csv")
print("- bias_audit_outcome_shift_summary.csv")
print("- bias_audit_top_features_<OUTCOME>.csv")
```
